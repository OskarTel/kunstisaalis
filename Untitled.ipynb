```python
from math import sqrt
#see funktsioon otsib üles keskmise värvi, aga võtab iga rgb osa ruutu ja siis lõpus leiab ruutjuure
def color_sq(rgb):
    r = 0
    g = 0
    b = 0
    for n in rgb:
        r += n[0]*n[0]
        g += n[1]*n[1]
        b += n[2]*n[2]
    r = round(sqrt(r/len(rgb)))
    g = round(sqrt(g/len(rgb)))
    b = round(sqrt(b/len(rgb)))
    return (r,g,b)

#see funktsioon otsib üles keskmise värvi, aga ei võta ruutu
def color(rgb):
    r = 0
    g = 0
    b = 0
    for n in rgb:
        r += n[0]
        g += n[1]
        b += n[2]
    r = round(r/len(rgb))
    g = round(g/len(rgb))
    b = round(b/len(rgb))
    return (r,g,b)
    
#see funktsioon ütleb eraldi ette antud rgb heleduse
def brightness0(r,g,b):
    hele = math.sqrt(0.299 * (r * r) + 0.587 * (g * g) + 0.114 * (b * b))
    if hele > 170:
        return 'light'
    elif hele > 85 and hele < 170:
        return 'medium'
    elif hele < 85:
        return 'dark'
    
def brightness(rgb):
    heledus = 0
    for n in rgb:
        heledus += math.sqrt(0.299 * (n[0] * n[0]) + 0.587 * (n[1] * n[1]) + 0.114 * (n[2] * n[2]))
    heledus = round(heledus / len(rgb))
    if heledus >= 170:
        return 'light'
    elif heledus > 85 and heledus < 170:
        return 'medium'
    elif heledus <= 85:
        return 'dark'

def rgb2gray(rgb):
    grayscaled = []
    for n in rgb:
        grayscaled.append([n[0] * 0.2989, n[1] * 0.5870, n[2] * 0.1140])
    return grayscaled



#def kontrast_kesk(ruudujarjend):

#Kui võtta arvesse eukleidese kaugust värvide eristamisel, siis proovime eristada põhilisi värve ja kui palju neid 
#antud ruudukeses on, et kas tegu on kirju ruudukesega või monotoonse ruudukesega
#def kirjudus(ruudujarjend):
#    colors = []
#    for i in range (len(ruudujarjend)-1):
#        distance = round(sqrt((ruudujarjend[i][0] - ruudujarjend[i+1][0])*(ruudujarjend[i][0] - ruudujarjend[i+1][0]) + (ruudujarjend[i][1] - ruudujarjend[i+1][1])*(ruudujarjend[i][1] - ruudujarjend[i+1][1]) + (ruudujarjend[i][2] - ruudujarjend[i+1][2])*(ruudujarjend[i][2] - ruudujarjend[i+1][2])))
#        if distance > 20:
#            colors.append(distance)
#    return colors
                                           
color([[2,4,333],[110,2,111],[203,0,222]])
#color_sq([[2,4,333],[110,2,111],[203,0,222]])
#heledus([[2,4,333],[110,2,111],[203,0,222]])
  #kirjudus([[5,5,5],[111,111,111],[0,0,0],[5,5,5],[111,111,111],[0,0,0]])
   #kontrast([[2,4,333],[110,2,111],[203,0,222]])
#rgb2gray([[2,4,333],[110,2,111],[203,0,222]])
brightness(rgb2gray([[2,4,333],[110,2,111],[203,0,222]]))

#def kontrast():
#    keskmine = 0
#    for n in ruudujarjend:
#        brightness = round((299*n[0] + 587*n[1] + 114*n[2]) / 1000)
#        keskmine += brightness
#    keskmine / len(ruudujarjend)
#    return keskmine
```
